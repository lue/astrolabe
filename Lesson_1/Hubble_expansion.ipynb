Some theory
===========

In 1929 [Edwin Hubble](http://en.wikipedia.org/wiki/Edwin_Hubble) made a conclusion about Universe expansion. In this class we will reproduce his analysis and will make another step forward and will talk about Dark Energy.

First, what do we mean by expansion? How can the space expand? The classic example is ants on the inflating balloon. For an observer (one ant) the position of other ants will be changing in time. He will notice that other objects will move away from him.

Question to think about: Why the ant himself will not expand in size? What will happen with the picture drawn on the balloon?

Now imagine ourselves on such a balloon, but now in 3D. How can we perform the same observations as an ant? We will need to understand two concepts.

First. Measuring the distance. An ant can use binocular vision to measure a distance. Can we do the same in astronomical observation? Not for cosmological objects. As you know from your daily life experience: the further the object — less effective your binocular vision to determine the distance to it. The cosmological objects are too far away to use this technique. We will return to it in another lesson, where we will observe nearby stars. So how can we do it? The [type Ia supernova](http://en.wikipedia.org/wiki/Type_Ia_supernova) is the answer. This is a type of star explosion with very narrow scatter of energy discharge. It allows to use those stars as standard candles. The distance to them can be measured from the flux we observe. Further the explosion — the fainter it looks

Second. Measuring the velocity. An ant measures the velocity by doing sequence of observations in time. But in our case it is useless since cosmological time scales are much longer than our life. Therefore, nothing visually changes on the night sky. Luckily, there is a way to measure velocity using Doppler effect. The light emitted by moving object is redshifted (or blueshifted) if it moves away (towards) us.

In this lesson, as well as in other lessons, we are not going to discuss physical aspects of the effects in details. There are tons of literature written on these topics for people with different backgrounds. We will focus only on the reproducing the numbers and statistical analysis.

Downloading the data:
---------------------

We will use data from this survey: http://supernova.lbl.gov/Union/

In particular we will need this dataset:

http://supernova.lbl.gov/Union/figures/SCPUnion2.1_mu_vs_z.txt

```python
import numpy as np
import matplotlib.pyplot as plt
import urllib2

#### Comment these two lines if you do not use iPython ####
from IPython.core.display import Image
%matplotlib inline
####

downloaded_data  = urllib2.urlopen('http://supernova.lbl.gov/Union/figures/SCPUnion2.1_mu_vs_z.txt')

# The header of the file contains five lines of description. We are skipping it.
# Also we are skipping first column which contains Supernova names, which has nontrivial format.
data=np.loadtxt(downloaded_data,skiprows=5,converters={0: lambda s: 0})
# Generating an array ‘z’ with redshifts.
z = data[:,1]
# Generating an array with distance modulus and it’s error.
DM = data[:,2]
DM_err=data[:,3]
```

What is distance modulus, you can quickly read on [Wikipedia](http://en.wikipedia.org/wiki/Distance_modulus). Let’s convert it into physical value using the equation provided in wiki article. Distance (luminosity distance) to the object we will put in array **DL**:

```python
# Defining conversion function.
def DM2DL(DM):
    return 10**(DM/5-1)/1e4
# Converting the whole array.
DL=DM2DL(DM)
```

The authors of the catalog assumed that the luminosity of the SNIa is -19.3. It is written in the catalog description, so we should not worry about this number. In fact, this number is important, but that’s another story.

Now we have redshifts in **z** and distances in **DL**. Redshift is unitless, while distances are in Mpc (1 pc = 3e16 m = 3.3 light years).

Hubble's idea was to plot one versus another and to see if there any correlation. Let's do the same:

```python
plt.plot(DL, z, '.')
plt.xlabel(r'$D_L\;\mathrm{[Mpc]}$',size=18)
plt.ylabel(r'$z$',size=18)
```

Great! There is a correlation. Now let's convert redshift, $z$, of objects into velocities, $v$.

Since the function converting $z \rightarrow v$ is not trivial, and $v \rightarrow z$ is very easy to implement, we will first write a function $v \rightarrow z$ and then make an interpolation between calculated points:

```python
# Defining speed of light
c=29979245800 # cm/s
# Function 
def v2z(v):
    # the equation is taken from http://en.wikipedia.org/wiki/Redshift#Redshift_formulae
    return np.sqrt((1.0+v/c)/(1.0-v/c))-1.0
```

For the reverse function we will need to first calculate velocity-redshift dependence in a series of points and then write a function which would  interpolate between them.

```python
v_list=np.linspace(0,c-1.,100)
z_list=v2z(v_list)

def z2v(z):
    return np.interp(z,z_list,v_list)/1e5 

v=z2v(z) # km/s
```

Now let's look how distance correlates with velocity:

```python
plt.plot(DL,v,'.')
plt.xlabel(r'$D_{L}\;\mathrm{[Mpc]}$',size=18)
plt.ylabel(r'$v\;\mathrm{[km/s]}$',size=18)
```

The figure clearly shows that the farther the object, higher the velocity it moves away from us.

Fitting the line
----------------

We can try to fit the line in this dependence. The slope the line (its first derivative) called Hubble constant, $H$, and has units $\rm{km/s/Mpc}$.

Now we are only going to estimate the fit by eye. Fitting the line using statistical methods will be your homework.

```python
plt.plot(DL, v, '.', label=r'$\mathrm{data}$')
plt.xlabel(r'$D_{L}\;\mathrm{[Mpc]}$', size=18)
plt.ylabel(r'$v\;\mathrm{[km/s]}$', size=18)
temp = np.array([0.0,6000.0])
plt.plot(temp, 50.0*temp,  'r', label='H=50')
plt.plot(temp, 70.0*temp,  'g', label='H=70')
plt.plot(temp, 100.0*temp, 'm', label='H=100')
plt.legend(loc=4)
```

Let's zoom in into the left bottom corner of the figure:

```python
plt.plot(DL, v, '.', label=r'$\mathrm{data}$')
plt.xlabel(r'$D_{L}\;\mathrm{[Mpc]}$', size=18)
plt.ylabel(r'$v\;\mathrm{[km/s]}$', size=18)
temp = np.array([0.0,6000.0])
plt.plot(temp, 50.0*temp,  'r', label='H=50')
plt.plot(temp, 70.0*temp,  'g', label='H=70')
plt.plot(temp, 100.0*temp, 'm', label='H=100')
plt.legend(loc=4)
plt.xlim(0, 250)
plt.ylim(0, 2e4)
```

The best guess out of three is $H=70\;\mathrm{km/s/Mpc}$, i. e. the object located at $2\;\mathrm{Mpc}$ from us is moving away with velocity $140\;\mathrm{km/s}$.

In other words, we can say that we were at the same point (assuming constant velocity):

$$\dfrac{2\;\mathrm{Mpc}}{140\;\mathrm{km/s}} = \dfrac{1}{H} = 14\,000\,000\,000\;\mathrm{yrs}$$

After a few lines of code we were able to estimate the age of the universe. It is 14 billion years old.

Firstly, we scipped the part with actual fitting and therefore we do not know the error of out measurment, which is the essential part of any scientific result. Secondly, we did not disscuss why the brightness of supernovae is constant.

Nevertheless, such a rough estimate gives pretty accurate result! More detailed analysis, including other measurment, gives the number very close to what we got: 13.813±0.058 billion years.

Dark Energy?
============

From figures above you may noticed that linear model fits only at the beginning of the diagram. ТNext step would be to include knowledge about cosmology. Let's consider simlest case, when the universe consists only from matter and dark energy. The fraction of the universe which consists of matter we will $\Omega_M$ and the dark energy $\Omega_L$ ($\Omega_M+\Omega_L=1$). You can refresh your knoledge about cosmology from here: [1](http://en.wikipedia.org/wiki/Distance_measures) [2](http://en.wikipedia.org/wiki/Friedmann%E2%80%93Lema%C3%AEtre%E2%80%93Robertson%E2%80%93Walker_metric). Luminosity distance, $d_L$, can be expressed like this:

$$d_L(z)  = (1+z) d_H \int_0^z \frac{dz'}{\sqrt{\Omega_m(1+z)^3+\Omega_\Lambda}}$$

Let's write the code, which calculates the integral:

```python
# Helping function E(z)
def e_Z(z, OmegaM):
    OmegaL=1.0-OmegaM
    return (OmegaM*(1.0+z)**3+OmegaL)**(-0.5)

# importing module for integrating
import scipy.integrate as si

# Function forcalculating luminosity distance 
def D_L(z, OmegaM):
    dh=4286.0 # это величина c/H, где H мы считаем равным 70 км/с/Мпс
    D_c=dh*si.quad(e_Z,0.0,z,args=(OmegaM))[0]
    return D_c*(1.0+z)

# TO make the function fork with arrays we write a wrapper arround it,
# or use @np.vectorize
def D_L_batch(z,OmegaM,H):
    DL=z.copy()
    for i in range(len(z)):
        DL[i]=D_L(z[i],OmegaM)
    return DL
```

Now let's plot the $d_L(z)$ dependence for a few values of $\Omega_M$:

```python
# Plotting the data
plt.plot(DL,v,'.')
plt.xlabel(r'$D_{L}\;\mathrm{[Mpc]}$',size=18)
plt.ylabel(r'$v\;\mathrm{[km/s]}$',size=18)
# temporary variables
temp_z=np.linspace(0,1.5,100)
temp_v=z2v(temp_z)
# adding models with different Omega_M
plt.plot(D_L_batch(temp_z,0.01,70.0),temp_v,'r')
plt.plot(D_L_batch(temp_z,0.30,70.0),temp_v,'g')
plt.plot(D_L_batch(temp_z,0.99,70.0),temp_v,'m')
plt.legend(('data',r'$\Omega_M=0.01$',r'$\Omega_M=0.30$',r'$\Omega_M=0.99$'),loc=4)
```

Green line seems to be the best fit. Therefore we may conclude that our universe consists of ~30% matter and ~70% dark energy. Those 30% of matter are not as simple as one may think. Most of it is 'dark matter'. We will explore evidence of it in later lessons.
